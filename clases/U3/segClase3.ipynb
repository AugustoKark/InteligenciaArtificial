```python
import numpy as np

seed = 22

n = 1000
umbral = 0.5
y_true = np.random.randint(0, 2, n)
y_pred = np.random.uniform(0, 1, n)

y_pred = (y_pred >= umbral).astype(int)
y_pred

```

```python
aciertos = np.sum(y_true == y_pred)
(aciertos/n) * 100

error= np.sum(y_true != y_pred)
(error)*100

(aciertos + error) *100
```
